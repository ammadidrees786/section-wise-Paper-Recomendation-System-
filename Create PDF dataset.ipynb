```python
import requests
from bs4 import BeautifulSoup
import os
import time
import zipfile
import shutil

def download_acl_pdfs_zip(zip_path='acl_dataset.zip', max_papers=3000):
    temp_dir = "acl_temp_pdfs"
    os.makedirs(temp_dir, exist_ok=True)
    base_url = "https://aclanthology.org"
    count = 0

    for year in range(2022, 2024):
        volume_url = f"https://aclanthology.org/events/acl-{year}/"
        try:
            resp = requests.get(volume_url)
            soup = BeautifulSoup(resp.text, "html.parser")
            paper_links = soup.find_all("a", href=True)

            for link in paper_links:
                href = link["href"]
                if href.startswith("/") and href.endswith("/"):
                    paper_id = href.strip("/")
                    pdf_url = f"https://aclanthology.org/{paper_id}.pdf"
                    try:
                        pdf_resp = requests.get(pdf_url)
                        if pdf_resp.status_code == 200:
                            file_path = os.path.join(temp_dir, f"{paper_id}.pdf")
                            with open(file_path, "wb") as f:
                                f.write(pdf_resp.content)
                            count += 1
                            print(f"Downloaded: {paper_id}")
                        else:
                            print(f"Skipped (not found): {pdf_url}")

                        if count >= max_papers:
                            print(f"\n✅ Downloaded {count} PDFs.")
                            raise StopIteration  # Break out of nested loop
                        time.sleep(1)
                    except Exception as e:
                        print(f"Error downloading {pdf_url}: {e}")
        except StopIteration:
            break
        except Exception as e:
            print(f"Failed to process year {year}: {e}")

    # Zip all downloaded files
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(temp_dir):
            for file in files:
                zipf.write(os.path.join(root, file), arcname=file)
    print(f"\n📦 Zipped {count} PDFs into {zip_path}")

    # Optional: remove the temp folder
    # shutil.rmtree(temp_dir)
    # print(f"🗑️ Temporary folder removed: {temp_dir}")

# Call the function
download_acl_pdfs_zip(zip_path='acl_dataset.zip', max_papers=3000)
```
